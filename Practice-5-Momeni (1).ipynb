```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_path = "dl2.jpg"

class ImageProcessor:
   def __init__(self, image_path):
        
        self.img = cv2.imread(img_path)
        self.img = cv2.cvtColor(self.img, cv2.COLOR_BGR2RGB)
        self.img = cv2.resize(self.img, (500,500))import cv2


    def get_grayscale(self,image):
        return cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

    def Process_red(self):
        img_red = self.img.copy()

        img_red[:,:,0] = np.clip(img_red[:,:,0] * 2,0, 255)
        return img_red

    def threshold_image(self, image):

        img_copy = image.copy()

        gray = self.get_grayscale(img_copy)
        img_copy[gray > 200] = [0,0,0]
        img_copy[gray < 50] = [255,255,255]
        return img_copy
        
    def invert_histogram(self,img):
        return 255 - image

    def run_all(self):
        img_red = self.process_red()
        img_threshold = self.threshold_image(img_red)
        img_inverted = self.invert_histogram(img_red)

        fig, axs = plt.subplots(4,2, figsize=(10,20))
        # ردیف 1
        axs[0,0].imshow(self.img)
        axs[0,0].set_title("Original Image")
        axs[0,1].hist(self.get_grayscale(img_red).ravel(), 256, [0, 256])
        axs[0,1].set_title("Histogram Original")

        axs[1,0].imshow(img_red)
        axs[1,0].set_title("IMG-RED")
        axs[1,1].hist(self.get_grayscale(img_threshold).ravel(), 256, [0, 256])
        axs[1,1].set_title("Histogram Thresholded")

        axs[2,0].imshow(img_threshold)
        axs[2,0].set_title("Thresholded")
        axs[2,1].hist(self.get_grayscale(img_threshold).ravel(), 256, [0,256])
        axs[2,1].set_title("Histogram Thresholded")

        axs[3,0].imshow(img_inverted)
        axs[3,0].set_title("Inverted Image")
        axs[3,1].hist(self.get_grayscale(img_inverted).ravel(), 256, [0,256])
        axs[3,1].set_title("Histogram Inverted")
      
        plt.tight_layout() 
        plt.show()
        obj = ImageProcessor("dl2.jpg")
        if obj.img is not None:
            obj.run_all()
        
```
